```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
path = r"C:\Users\Shenghui\Documents\GitHub\MSc_DV_CC\Cryptocurrency Transaction Data.csv"
```

```python
df = pd.read_csv(path)
df.head()
```

```python
df.isnull().sum()
```

```python
bit_df = df[df["Currency"]=="BTC"]
bit_df.head()
```

```python
total = len(bit_df)
Max_amount = bit_df["Amount"].max()
Min_amount = bit_df["Amount"].min()
Max_T = bit_df["Timestamp"].max()
Min_T = bit_df["Timestamp"].min()
print(f"{Max_amount=:}")
print(f"{Min_amount=:}")
print(f"{Max_T=:}")
print(f"{Min_T=:}")
print("Total transcations:", total)
```

```python
counts = bit_df["Receiver_Address"].value_counts()
duplicates = counts[counts > 1].index.tolist()
print(duplicates)

counts = bit_df["Sender_Address"].value_counts()
duplicates = counts[counts > 1].index.tolist()
print(duplicates)

#c = pd.Series(["a","a"])
counts = bit_df["Block_ID"].value_counts()
duplicates = counts[counts > 1].index.tolist()
print(duplicates)
```

```python
gradient = np.linspace(math.floor(Min_amount), math.ceil(Max_amount)).reshape(1, -1)

plt.figure(figsize=(8, 1.5))
plt.imshow(gradient, aspect='auto', cmap='rainbow') 
plt.yticks([])
plt.show()
```

```python
G = nx.Graph()
edges = list(zip(df["Sender_Address"][:500],df["Receiver_Address"][:500]))
G.add_edges_from(edges)
plt.figure(figsize=(12, 12))
nx.draw_spring(G, node_size=10, edge_color="gray", alpha=0.3)
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_edge(1, 2)
G.add_edge(1, 3)
G.add_edge(1, 5)
G.add_edge(2, 3)
G.add_edge(3, 4)
G.add_edge(4, 5)

# explicitly set positions
pos = {1: (0, 0), 2: (-1, 0.3), 3: (2, 0.17), 4: (4, 0.255), 5: (5, 0.03)}

options = {
    "font_size": 36,
    "node_size": 3000,
    "node_color": "green",
    "edgecolors": "black",
    "linewidths": 5,
    "width": 5,
}
nx.draw_networkx(G, pos, **options)

# Set margins for the axes so that nodes aren't clipped
ax = plt.gca()
ax.margins(0.20)
plt.axis("off")
plt.show()
```

```python
G = nx.DiGraph([(0, 3), (1, 3), (2, 4), (3, 5), (3, 6), (4, 6), (5, 6)])

# group nodes by column
left_nodes = [0, 1, 2]
middle_nodes = [3, 4]
right_nodes = [5, 6]

# set the position according to column (x-coord)
pos = {n: (0, i) for i, n in enumerate(left_nodes)}
pos.update({n: (1, i + 0.5) for i, n in enumerate(middle_nodes)})
pos.update({n: (2, i + 0.5) for i, n in enumerate(right_nodes)})

nx.draw_networkx(G, pos, **options)

# Set margins for the axes so that nodes aren't clipped
ax = plt.gca()
ax.margins(0.20)
plt.axis("off")
plt.show()
```
